```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1.0/255) # Initialize the train_datagen with the value returned from ImageDataGenerator
val_datagen = ImageDataGenerator(rescale=1.0/255) # Initialize the val_datagen with the value returned from ImageDataGenerator
```

```python
train_generator = train_datagen.flow_from_directory('../input/horse-or-human/horse-or-human/train',
                                                    target_size=(300, 300),
                                                    batch_size=32,
                                                    class_mode='binary')
# Set the train_generator and use the training dataset from the data.
```

```python
val_generator = val_datagen.flow_from_directory('../input/horse-or-human/horse-or-human/validation',
                                                target_size=(300, 300),
                                                batch_size=32,
                                                class_mode='binary')
# Set val_generator to the validation data from the dataset
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3, 3), activation=tf.nn.relu,
                           input_shape=(300, 300, 3)),
    tf.keras.layers.MaxPool2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(32, (3, 3), activation=tf.nn.relu),
    tf.keras.layers.MaxPool2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation=tf.nn.relu),
    tf.keras.layers.MaxPool2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation=tf.nn.relu),
    tf.keras.layers.MaxPool2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation=tf.nn.relu),
    tf.keras.layers.MaxPool2D(pool_size=(2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=512, activation=tf.nn.relu),
    tf.keras.layers.Dense(units=1, activation=tf.nn.sigmoid)
])

# Build the keras model
# takes as argument a list of layers added to the model
# This is where we define our neural network
# This model will further be used for validation in the next step
```

```python
model.compile(optimizer=tf.train.AdamOptimizer(),
              loss='binary_crossentropy',
              metrics=['accuracy'])
# Compile the created model 
```

```python
model.fit_generator(train_generator,
                    steps_per_epoch=32,
                    epochs=20,
                    validation_data=val_generator,
validation_steps=8)

# Use the fit generator method to test out model with 20 epochs and using the validation data from the dataset for validation.
```
